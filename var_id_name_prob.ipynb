```python
def sum_digit(id):
    sum=0
    for digit in str(id):
        sum+=int(digit)
    return sum
    

output=''
name,id=input().split()
id_list=list(id)
id=int(id)
sum=sum_digit(id)
if(name[0].upper()==name[-1].upper()):
    output+=str(len(name))

if sum%2==0:
    output+=str(sum)+str('E')
if id_list[-1]==id_list[-2]:
    output+=str(id_list[-1])
    
if name.count('a')>1:
    output+=str('a')
if name.startswith('A') and id_list[-1]==9:
    output+=str('A9')
    
if len(output)==0:
    print('X')

print(" Output_string is : {}".format(output))
```
